```python
WINDOW_SIZE = 30
```

```python
def create_sequences(df, window_size, feature_cols, target_col):
    X, y = [], []

    for engine_id in df['engine_id'].unique():
        engine_df = df[df['engine_id'] == engine_id]

        features = engine_df[feature_cols].values
        targets  = engine_df[target_col].values

        for i in range(len(engine_df) - window_size + 1):
            X.append(features[i:i+window_size])
            y.append(targets[i+window_size-1])

    return np.array(X), np.array(y)
```

```python
X_train_seq, y_train_seq = create_sequences(
    train_data_scaled,
    WINDOW_SIZE,
    feature_cols,
    TARGET_COL
)

X_val_seq, y_val_seq = create_sequences(
    val_data_scaled,
    WINDOW_SIZE,
    feature_cols,
    TARGET_COL
)
```

```python
def create_test_sequences(df, window_size, feature_cols):
    X = []
    engine_ids = []

    for engine_id in df['engine_id'].unique():
        engine_df = df[df['engine_id'] == engine_id]

        features = engine_df[feature_cols].values

        X.append(features[-window_size:])
        engine_ids.append(engine_id)

    return np.array(X), engine_ids
```

```python
# defining the baseline model 
import torch.nn as nn

class RUL_LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, dropout):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0
        )

        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        return self.fc(out).squeeze(1)
    
```

```python
import optuna
import math
```

```python
def objective(trail):
    input_size = len(feature_cols)
    window_size = trail.suggest_int("window_size",20,50)
    hidden_size = trail.suggest_int("hidden_size",32,256)
    num_hidden_layers = trail.suggest_int("num_hidden_layers",1,3)
    dropout  = trail.suggest_float("dropout",0.0,0.50)
    lr          = trail.suggest_float("lr", 1e-4, 1e-2, log=True)
    batch_size = trail.suggest_categorical("batch_size", [32, 64, 128])
    X_train_seq, y_train_seq = create_sequences(
    train_data_scaled,
     window_size,
    feature_cols,
    TARGET_COL
             )

    X_val_seq, y_val_seq = create_sequences(
         val_data_scaled,
          window_size,
         feature_cols,
         TARGET_COL
            )

    train_dataset = TensorDataset(
        torch.tensor(X_train_seq, dtype=torch.float32),
        torch.tensor(y_train_seq, dtype=torch.float32)
    )
    
    val_dataset = TensorDataset(
        torch.tensor(X_val_seq, dtype=torch.float32),
        torch.tensor(y_val_seq, dtype=torch.float32)
    )
    

    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True
    )
    
    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False
    )
    
         #self,, input_size, hidden_size, num_layers, dropout
    model = RUL_LSTM(input_size,hidden_size,num_hidden_layers,dropout)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.MSELoss()
    epochs = 10 
    for epoch in range(epochs):
      model.train()
      for X,y in train_loader :
          y_pred = model(X)
          loss = loss_fn(y_pred,y)
          optimizer.zero_grad()
          loss.backward()
          optimizer.step()
    total_loss = 0
    model.eval()
    with torch.no_grad():
      for X,y in val_loader :
          y_pred = model(X)
          loss = loss_fn(y_pred,y)
          total_loss+=loss.item()
    return math.sqrt(total_loss/len(val_loader))
        
          
        
```

```python
# now finally lets create the study 
study = optuna.create_study(direction = "minimize")
study.optimize(objective, n_trials=30)
print("Best validation RMSE:", study.best_value)
print("Best hyperparameters:", study.best_params)
```

```python
best = study.best_params

X_train_all, y_train_all = create_sequences(
    pd.concat([train_data_scaled, val_data_scaled]),
    best["window_size"],
    feature_cols,
    TARGET_COL
)

train_all_ds = TensorDataset(
    torch.tensor(X_train_all, dtype=torch.float32),
    torch.tensor(y_train_all, dtype=torch.float32)
)

train_all_loader = DataLoader(
    train_all_ds,
    batch_size=best["batch_size"],
    shuffle=True
)
```

```python
model = RUL_LSTM(
    input_size=len(feature_cols),
    hidden_size=best["hidden_size"],
    num_layers=best["num_hidden_layers"],
    dropout=best["dropout"]
)

optimizer = torch.optim.Adam(model.parameters(), lr=best["lr"])
criterion = nn.MSELoss()

EPOCHS = 60

for epoch in range(EPOCHS):
    model.train()
    for xb, yb in train_all_loader:

        optimizer.zero_grad()
        loss = criterion(model(xb), yb)
        loss.backward()
        optimizer.step()
```
